Load the dataset sales_data.csv into a DataFrame.
Group the data by Region and calculate the total Sales for each region.
Group by Region and Product, then calculate the total Sales and average Units.
Group by Product and calculate multiple aggregations for Sales and Units

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Load the dataset sales_data.csv into a DataFrame
sales = pd.read_csv('sales_data.csv')
sales.head()
```

```python
# Group the data by Region
sale_region_groups = sales.groupby(sales['Region'])
sale_region_groups.describe()
```

```python
# calculate the total Sales for each region
sale_region_groups['Sales'].sum()
```

```python
# Group by Region and Product
sale_region_product_groups = sales.groupby(['Region', 'Product'])
sale_region_product_groups.describe()
```

```python
# calculate the total Sales
sale_region_product_groups['Sales'].sum()
```

```python
# calculate the average Units
sale_region_product_groups['Units'].mean()
```

```python
# Group by Product
sale_product_groups = sales.groupby(['Product'])
sale_product_groups.describe()
```

```python
# calculate multiple aggregations for Sales
sale_product_groups['Sales'].agg(['sum', 'mean'])
```

```python
# calculate multiple aggregations for Units
sale_product_groups['Units'].agg(['min', 'max'])
```
