## Extreme value analysis using univariate methods

**Definition**: An outlier is a data point significantly different from others in a dataset.

**Types of Outliers:**
Univariate outliers: Deviations in a single variable.
Multivariate outliers: Deviations in a combination of variables.

**Causes**:
Data entry errors.
Measurement errors.
Genuine extreme values.

**Why Detect Outliers?**
They can distort statistical results and predictions.
