```python
import numpy as np
import pandas as pd
```

```python
# a.	Generate a two-dimensional integer array with the shape of (20,5) and name its “data”.
#     Fill “data” with random integers from -100 to 100.
#     Display “data” at the end.

data = np.random.randint(-100,100,(20,5))
data
```

```python
# b.	Display the average value of the third column in “data”.
data[:,2].mean()
```

```python
# c.	Display the average value of the first 5 rows in “data”.
data[:5,:].mean()
```

```python
# d.	Display the average value of the first 3 rows and the 4th column in “data”.
data[:3,3].mean()
```

```python
# e.	Create a new NumPy array which contains the last 4 elements of the second row of “data”.  Display the new array at the end.
new_np_array = data[1,1:]
new_np_array
```

```python
# f.	Reshape “data” int to 10 x 10 matrix.  Display the matrix at the end.
data_reshaped = data.reshape(10, 10)
data_reshaped
```

```python
# g.	Calculate the std deviation of “data”.
data_reshaped.std()
```

```python
# h.	Replace the largest number of each row in “data” with the average of the corresponding row.  Display “data” at the end.

# Calculate the average of each row
row_averages = np.mean(data_reshaped, axis=1)

# Find the index of the maximum value in each row
max_indices = np.argmax(data_reshaped, axis=1)

# Replace the maximum values with the row averages
data_reshaped[np.arange(data_reshaped.shape[0]), max_indices] = row_averages

data_reshaped
```

```python
# i.	Create a new NumPy array which contains random 2 rows in “data”.  Display the new array at the end.
new_array_2 = data_reshaped[np.random.choice(data_reshaped.shape[0], size=2, replace=False)]
new_array_2
```

```python
# j.	Create a new NumPy array which contains random 3 columns in “data”.  Display the new array at the end.
new_array_3 = data_reshaped[:, np.random.choice(data_reshaped.shape[1], size=3, replace=False)]
new_array_3
```

```python
# k.	Create a new NumPy array which contains 100 random doubles drawn from a normal (Gaussian) distribution.  Display the new array at the end.
new_array_4 = np.random.randn(100)
new_array_4
```

```python
# l.	Convert the "data" to a Pandas Data Frame with the header “a” – “e” and name it “dataframe”. Describe the “dataframe” at the end.
dataframe = pd.DataFrame(data, columns=['a', 'b', 'c', 'd', 'e'])
dataframe.describe()
```

```python
# n.	Save the “dataframe” into assign1.csv.
dataframe.to_csv('assign1.csv', index=False)
```
