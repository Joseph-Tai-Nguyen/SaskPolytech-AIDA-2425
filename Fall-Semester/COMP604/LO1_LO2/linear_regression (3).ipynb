<h2 style="color:green" align="center">Machine Learning With Python: Linear Regression With One Variable</h2>

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

# homeprice = m*area + C

```python
df = pd.read_csv('homeprices.csv')
df
```

what is price of home if we have 5000 sqm

```python
%matplotlib inline
plt.xlabel('area')  # independent variable
plt.ylabel('price')  #dependent variable
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
#seperating the x/independent variable
new_df = df.drop('price',axis='columns')
new_df  #input
```

```python
price = df.price  #y /dependent variable
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)  #.fit =>training (input,output)
```

**(1) Predict price of a home with area = 3300 sqr ft**

```python
reg.predict([[3300]])
```

```python
reg.coef_   #slope of line
```

```python
reg.intercept_   #y-axis intercept
```
